```python
%%writefile -a app.py
# --- Solution: Part 3 - Add New Book Form ---
st.subheader("Add a New Book")
with st.form("add_book_form"):
    title = st.text_input("Title")
    author = st.text_input("Author")
    genre = st.text_input("Genre")  # could be a selectbox for fixed genres
    pages = st.number_input("Pages", min_value=1, step=1, value=100)
    read = st.checkbox("Read")
    submitted = st.form_submit_button("Add Book")
    if submitted:
        # validation
        if not title.strip():
            st.error("Title cannot be empty. Please enter a title.")
        elif not author.strip():
            st.error("Author cannot be empty. Please enter an author.")
        elif pages <= 0:
            st.error("Pages must be a positive integer.")
        else:
            # append to session_state dataframe
            new_row = {"Title": title.strip(), "Author": author.strip(), "Genre": genre.strip(), "Pages": int(pages), "Read": bool(read)}
            if "books_df" not in st.session_state or st.session_state["books_df"] is None:
                st.session_state["books_df"] = pd.DataFrame([new_row])
            else:
                st.session_state["books_df"] = pd.concat([st.session_state["books_df"], pd.DataFrame([new_row])], ignore_index=True)
            st.success("Book added successfully.")
            # optional: rerun to refresh UI in some cases (not necessary in notebook)
            # st.experimental_rerun()
            # After form, display updated table
st.subheader("Updated Book Library")
st.dataframe(st.session_state.get("books_df", pd.DataFrame()))
```
