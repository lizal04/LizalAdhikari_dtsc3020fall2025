## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

```python
# TODO: Create the scatter plot here
p = sns.scatterplot(x=penguins.bill_length_mm, y= penguins.bill_depth_mm, hue = penguins.species)
p.set_xlabel("Bill length (mm)")
p.set_ylabel("Bill depth (mm)")
p.set_title("Penguin Bill Comparion across species")
```
